```python
#Import Dependencies
import pandas as pd
```

```python
#Read Data
path = "./Resources/loan_approval_dataset.csv"
data = pd.read_csv(path)
data = data.dropna()
data[:5]
```

```python
#Factor down Financial Values by 100
def div100(num):
    return (num/100).round()

data.iloc[:, [4,5,8,9,10,11]] = data.iloc[:, [4,5,8,9,10,11]].apply(div100).astype(int)
data
```

```python
#Exporting Sheet as CSV
data.to_csv("Resources/loan_data.csv")
```
